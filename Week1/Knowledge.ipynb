### Model Checking Algorithms
How to set the entailment relationship between KB and a?:
    1. Enumerate all possibile models (for loop)
    2. If in every model where KB is true, a is also true, then KB entails a.

In pain Text: Model Checking Algotithms takes a KB and a query, go through all the possible "world" (models with different truth/falsh values for each relevant symbols), to exam if in every model that KB is ture also return a true value for query, then the entailment exists.

This is also called Knowledge Engineering: take real world knowledge turn it into eintailment pairs for the computer to draw conclussion on certain queries

### Inference Rules Algorithms

#### Mudus Ponens
Modus ponens is another approach, which is: 
when a implicate b AND a, then b.

#### And Elimination
a AND b -> a true or b true

#### Double Negation Elimination
Not not a -> a true

#### Implication Elimination
a -> b, then not a or b

#### Biconditional Elimination
a <-> b
a->b AND a<-b 

#### De Morgan Law
Not (a AND B), then (Not a) or (Not b)



### Themorem Proving

We can use the previously listed inference Rlues into a search Problem:
1. initial state: starting knowledge base
2. actions: inference rules
3. transition model: new knowledge base after inference
4. goal test: check statement we're trying to prove
5. path cost function: number of steps in proof
