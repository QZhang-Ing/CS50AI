## Introduce Probability to AI

Possible Worlds(Values): w

Probability of w: P(w)

Sum(P(w)) = 1


### Conditional Probability
Probability of given condition:
P(a|b): probability of a given b

P(rain today | rain yesterday)
P(disease | test results)

How to calculate:

P(a|b) = P(a ^ b) / P(b)
P(a ^ b) = P(b) * P(a|b)
P(a ^ b) = P(a) * P(b|a)

#### Independent 
if a and b are independent form each other:
P(a|b) = P(a) * P(b)

## Bayes' Rule

P(b|a) = P(a|b) * P(b) / P(a)

## Joint Probability


## Inclusion-Exclusion
P(a or b) = P(a) + P(b) - P(a, b)

### Inference

A inference:

1. Query X: the variable that we want to compute the distribution
